```python
import pandas as pd

import numpy as np
pd.set_option('display.max_columns', None)
noms_colonnes = [
    'modele', 'Version', 'Prix_en_euros', 'Durée_publi_en_j', 'volume_coffre_L', 'Options', 'Informations_techniques', 'Equipements'
]

df = pd.read_csv('thread2.csv', names=noms_colonnes)

# Afficher les différentes valeurs pour chaque colonne
for colonne in df.columns:
    print(f"Colonnes: {colonne}")
    print(df[colonne].unique())
    print("\n")
```

```python
df.head()
```

```python
# Ajoutez les colonnes "marque" et "modele"
df[['marque', 'modele']] = df['modele'].str.split(n=1, expand=True)

# Affichez le DataFrame résultant
print(df)
```

```python
# Divisez la colonne "Version" en plusieurs colonnes
df[['cylindree_moteur', 'type_moteur', 'puissance', 'finition']] = df['Version'].str.strip("()").str.replace("'", '').str.split(", ", expand=True)
df.drop('Version', axis=1, inplace=True)
df[['cylindree_moteur', 'type_moteur', 'puissance', 'finition']] = df[['cylindree_moteur', 'type_moteur', 'puissance', 'finition']].apply(lambda x: x.str.replace("'", ''))


# Affichez le DataFrame résultant
print(df)
```

```python
# Nettoyer la colonne "Prix" en ne conservant que les valeurs numériques
# Supprimer les caractères non numériques de la colonne "Prix"
df['Prix_en_euros'] = df['Prix_en_euros'].replace('[^\d.]', '', regex=True)
df['Prix_en_euros'] = df['Prix_en_euros'].str.replace(' €', '').str.replace('\u202f', '').astype(int)

# Afficher le DataFrame pour vérifier les modifications
print(df)

```

```python
# Nettoyer la colonne "Durée_publi" en ne conservant que les valeurs numériques
df['Durée_publi_en_j'] = df['Durée_publi_en_j'].str.extract('(\d+)').fillna(0).astype(int)

# Remplacer les valeurs manquantes par la moyenne arrondie vers le haut
mean_value = np.ceil(df['Durée_publi_en_j'].mean())
df['Durée_publi_en_j'] = df['Durée_publi_en_j'].replace({0: mean_value}).astype(int)
```

```python
# Diviser la colonne "Options" en plusieurs colonnes
options_columns = df['Options'].apply(eval).apply(pd.Series)

# Renommer les colonnes en option_1, option_2, ..., option_8
options_columns.columns = [f'option_{i+1}' for i in range(options_columns.shape[1])]

# Nettoyer les valeurs des colonnes d'options
options_columns = options_columns.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Ajouter les nouvelles colonnes d'options au DataFrame principal
df = pd.concat([df, options_columns], axis=1)
# Supprimer la colonne "Options"
df = df.drop('Options', axis=1)
# Afficher le DataFrame pour vérifier les modifications
print(df)
```

```python
# Supprimer les caractères indésirables
df['Informations_techniques'] = df['Informations_techniques'].str.replace("\n", "").str.replace("\r", "")

# Extraire chaque paire clé-valeur pour chaque élément du DataFrame
pattern = r"\('(.*?)', '(.*?)'\)"
df_extracted = df['Informations_techniques'].str.findall(pattern).apply(lambda x: dict(x))

# Créer une colonne pour chaque nom de clé
for col_name in set(col for d in df_extracted for col in d.keys()):
    df[col_name] = df_extracted.apply(lambda x, col=col_name: x.get(col, None))

# Afficher le DataFrame résultant
print(df)
```
