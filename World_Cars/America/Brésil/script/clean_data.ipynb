```python
import pandas as pd
import glob

# Chemin du dossier contenant les fichiers CSV à fusionner
dossier_csv = r'E:\GITHUB\Cars_Analysis\World_Cars\America\Brésil\script'

# Modèle de nom de fichier (par exemple, tous les fichiers avec l'extension .csv)
modele_fichier = '*.csv'

# Liste pour stocker les DataFrames de chaque fichier
list_df = []

# Lire tous les fichiers CSV dans le dossier spécifié
for fichier_csv in glob.glob(dossier_csv + "\\" + modele_fichier):
    df = pd.read_csv(fichier_csv)
    list_df.append(df)

# Vérifier les fichiers trouvés
print("Fichiers trouvés :", glob.glob(dossier_csv + "\\" + modele_fichier))

# Vérifier si la liste des DataFrames est vide
if not list_df:
    print("Aucun fichier CSV trouvé dans le dossier spécifié.")
else:
    # Fusionner les DataFrames en un seul
    df_merged = pd.concat(list_df, ignore_index=True)

    # Afficher toutes les colonnes sans tronquer
    pd.set_option('display.max_columns', None)

    # Supprimer les colonnes
    colonnes_a_supprimer = ['Car Features']
    df_merged = df_merged.drop(colonnes_a_supprimer, axis=1)

    # Enregistrer le fichier fusionné
    fichier_fusionne2 = r'E:\GITHUB\Cars_Analysis\World_Cars\America\Brésil\script\fichier_nettoye2.csv'
    df_merged.to_csv(fichier_fusionne2, index=False)

    print("Fusion terminée. Fichier enregistré à", fichier_fusionne2)

# Afficher toutes les lignes du DataFrame
pd.set_option('display.max_rows', None)

# Charger le fichier fusionné
df_merged = pd.read_csv(fichier_fusionne2)

# Afficher les noms de colonnes disponibles
print("Noms de colonnes disponibles :", df_merged.columns)

# Afficher toutes les lignes du DataFrame
df_merged
```
