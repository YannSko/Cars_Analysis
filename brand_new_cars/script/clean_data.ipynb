```python
import pandas as pd
import glob

# Chemin du dossier contenant les fichiers CSV à fusionner
dossier_csv = r'E:\GITHUB\Cars_Analysis\brand_new_cars\script\CSV\Voiture neuve csv'

# Modèle de nom de fichier (par exemple, tous les fichiers avec l'extension .csv)
modele_fichier = '*.csv'

# Liste pour stocker les DataFrames de chaque fichier
list_df = []

# Lire tous les fichiers CSV dans le dossier spécifié
for fichier_csv in glob.glob(dossier_csv + "\\" + modele_fichier):
    df = pd.read_csv(fichier_csv)
    list_df.append(df)

# Vérifier les fichiers trouvés
print("Fichiers trouvés :", glob.glob(dossier_csv + "\\" + modele_fichier))

# Vérifier si la liste des DataFrames est vide
if not list_df:
    print("Aucun fichier CSV trouvé dans le dossier spécifié.")
else:
    # Fusionner les DataFrames en un seul
    df_merged = pd.concat(list_df, ignore_index=True)
    

    # Sélectionner les colonnes numériques
    colonnes_numeriques = df_merged.select_dtypes(include='number').columns

    # Supprimer les liens https de la colonne "lien_colonne"
    colonne_lien = "Lien_du_titre"  # Remplacez cela par le nom réel de votre colonne
    df_merged[colonne_lien] = df_merged[colonne_lien].str.replace(r'https://', '', regex=True)

    # Liste des colonnes à fusionner
    colonnes_a_fusionner = ['Champ1', 'Titre', 'ncisearchproduct_model', 'rbold','Modèle']  # Remplacez cela par les noms réels de vos colonnes

    # Créer une nouvelle colonne avec la concaténation des valeurs des colonnes sélectionnées
    df_merged['car_name'] = df_merged[colonnes_a_fusionner].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)

    # Déplacer la colonne 'car_name' en première position
    car_name = df_merged.pop('car_name')
    df_merged.insert(0, 'car_name', car_name)


    # Liste des colonnes à fusionner
    colonnes_a_fusionner2 = ['Champ2','fcaboxwrapper','audimodelfinder_snapelement','prix','étiquette5','image','customcheckbox','champ','modelfinder_item'	,'prix3',	'Lien_du_titre'	,'information',	'prix1'	,'motclé',	'prix2',	'prix4',	'étiquette'	,'promotionaltext'	,'amount'	,'scjjtsdx',	'description',	'description5','Titre1','Lien_du_titre1','image1','information1','prix5'	,'pri1'	,'motclé1',	'pri2',	'pri3'	,'pri4'	,'étiquette1',	'promotionaltext1'	,'amount1'	,'scjjtsdx1',	'description1'	,'descriptio1'	,'Titre2'	,'Lien_du_titre2'	,'image2',	'information2',	'prix6'	,'pri5',	'motclé2',	'pri6'	,'pri7',	'pri8',	'étiquette2',	'promotionaltext2',	'amount2',	'scccvcax'	,'description2',	'descriptio2','Prix1','Prix','col121','count','bt3']  # Remplacez cela par les noms réels de vos colonnes

    # Créer une nouvelle colonne avec la concaténation des valeurs des colonnes sélectionnées
    df_merged['car_price'] = df_merged[colonnes_a_fusionner2].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)

    # Déplacer la colonne 'car_price' en première position
    car_price = df_merged.pop('car_price')
    df_merged.insert(1, 'car_price', car_price)

    # Supprimer les colonnes d'origine
    df_merged = df_merged.drop(colonnes_a_fusionner2, axis=1)

    # Supprimer les colonnes 'colonne1' et 'colonne2'
    colonnes_a_supprimer = ['Champ1', 'Titre','ncisearchproduct_model','rbold','Modèle']
    df_merged = df_merged.drop(colonnes_a_supprimer, axis=1)

    # Afficher toutes les colonnes sans tronquer
    pd.set_option('display.max_columns', None)

    # Enregistrer le fichier fusionné
    fichier_fusionne = r'E:\GITHUB\Cars_Analysis\brand_new_cars\script\fichier_nettoye.csv'
    df_merged.to_csv(fichier_fusionne, index=False)

    print("Fusion terminée. Fichier enregistré à", fichier_fusionne)

# Afficher toutes les lignes du DataFrame
pd.set_option('display.max_rows', None)

# Charger le fichier fusionné
df_merged = pd.read_csv(fichier_fusionne)

# Afficher les noms de colonnes disponibles
print("Noms de colonnes disponibles :", df_merged.columns)

# Afficher toutes les lignes du DataFrame
df_merged
```
